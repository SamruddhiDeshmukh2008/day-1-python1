```python
count = 12345
add = 0
while count <= 5:
    count = count % 10
    count = count // 10
    count += 1
    print("addition is", add)
```

```python
count = 12345
add = 0
while count <= 6:
    rem = count % 10
    add = add + rem
    count = count // 10

    print("addition is", add)
```

```python
# break, continue
```

```python
for i in range(1, 10):
    if i == 5:
       break
    print(i)
    print("ok")
```

```python
for i in range(1 , 10):
    print(i)
    print("ok")
    if i == 5:
        break
```

```python
for i in range(1, 10):
    if i == 5:
       continue
    print(i)
```

```python
for i in range (1 , 11):
    print(i)
    if i == 5:
        continue
    print("ok")
    
```
